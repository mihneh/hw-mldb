### Теперь подготовим структуру каталогов для обучения YOLO. При формировании train и test нужно стратифицировать данные по классам ввиду дизбаланса.

```python
import os
from collections import defaultdict
import numpy as np
import shutil

# Функция для получения классоф фреймов
def get_image_classes(frames_dir):
    image_classes = defaultdict(set)
    for filename in os.listdir(frames_dir):
        if filename.endswith('.txt') and filename != 'classes.txt':
            annot_path = os.path.join(frames_dir, filename)
            with open(annot_path, 'r') as f:
                lines = f.readlines()
                if lines:
                    for line in lines:
                        if line.strip():
                            class_id = line.strip().split()[0]
                            image_name = os.path.splitext(filename)[0] + '.jpg'
                            image_classes[image_name].add(class_id)
                else:
                    # Если файл аннотации пустой
                    image_name = os.path.splitext(filename)[0] + '.jpg'
                    image_classes[image_name].add('no_object')
    return image_classes
```

```python
from sklearn.model_selection import train_test_split

# Функция для стратифицированного деления на train/test
def stratified_split(image_classes, test_size=0.2, random_state=42):
    images = list(image_classes.keys())
    labels = []
    for classes in image_classes.values():
        labels.append('_'.join(sorted(classes)))
    X_train, X_val, y_train, y_val = train_test_split(images,
                                                      labels,
                                                      test_size=test_size,
                                                      random_state=random_state,
                                                      stratify=labels)
    return X_train, X_val, y_train, y_val
```

### Теперь организуем файловую структуру

```python
def organize_files(frames_dir, X_train, X_val):
    # Создаем структуру папок
    for split in ['train', 'val']:
        images_dir = os.path.join(frames_dir, split, 'images')
        labels_dir = os.path.join(frames_dir, split, 'labels')
        os.makedirs(images_dir, exist_ok=True)
        os.makedirs(labels_dir, exist_ok=True)
    
    # Функция для перемещения файлов с проверкой существования
    def move_files(image_names, split):
        for image_name in image_names:
            # Путь к исходным файлам
            image_src = os.path.join(frames_dir, image_name)
            label_src = os.path.join(frames_dir, os.path.splitext(image_name)[0] + '.txt')
            # Путь к целевым папкам
            image_dst = os.path.join(frames_dir, split, 'images', image_name)
            label_dst = os.path.join(frames_dir, split, 'labels', os.path.splitext(image_name)[0] + '.txt')
            # Проверяем существование файлов перед перемещением
            if os.path.exists(image_src) and os.path.exists(label_src):
                shutil.move(image_src, image_dst)
                shutil.move(label_src, label_dst)
            else:
                print(f"Отсутствует изображение или аннотация для {image_name} в {split}.")

    # Перемещаем файлы
    move_files(X_train, 'train')
    move_files(X_val, 'val')
```

```python
import os
import cv2
import pandas as pd
from ultralytics import YOLO
```

```python
# Функция для извлечения кадров с видео
def extract_frames(video_path, step=0.5):
    cap = cv2.VideoCapture(video_path)
    fps = cap.get(cv2.CAP_PROP_FPS)
    if fps == 0 or fps is None:
        print(f"Не удалось получить FPS для {video_path}. Пропуск файла.")
        return []
    frame_interval = int(fps * step)
    frames = []
    frame_count = 0

    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            break
        if frame_count % frame_interval == 0:
            timestamp = frame_count / fps
            frames.append((timestamp, frame))
        frame_count += 1

    cap.release()
    return frames
```

```python
# Функция для инференса на списке кадров
def perform_inference(frames, model, video_file_name, date_folder):
    results = []
    for idx, (timestamp, frame) in enumerate(frames):
        res = model.predict(frame)
        result = res[0]
        if result.boxes:
            boxes = result.boxes
            for box in boxes:
                class_id = int(box.cls.item())
                bbox = box.xyxy.tolist()[0]
                state = 'broken' if class_id == 1 else 'whole'

                results.append({
                    'date': date_folder,
                    'filename': video_file_name,
                    'timing(sec)': timestamp,
                    'location': bbox,
                    'state': state
                })
        else:
            results.append({
                'date': date_folder,
                'filename': video_file_name,
                'timing(sec)': timestamp,
                'location': None,
                'state': 'no target'
            })
    return results
```

```python
# Функция для обработки одного видео
def process_video(video_path, model, csv_file_path, step=0.5):
    video_file_name = os.path.basename(video_path)
    date_folder = os.path.basename(os.path.dirname(video_path))
    
    frames = extract_frames(video_path, step=step)
    if not frames:
        print(f"Кадры не были извлечены из видео: {video_file_name}")
        return
    
    results = perform_inference(frames, model, video_file_name, date_folder)
    if results:
        df = pd.DataFrame(results)
        if os.path.exists(csv_file_path):
            df.to_csv(csv_file_path, mode='a', index=False, header=False)
        else:
            df.to_csv(csv_file_path, mode='w', index=False, header=True)
    else:
        print(f"Не удалось получить результат для видео: {video_file_name}")
```

```python
# Функция для обработки всех видео
def process_all_videos(videos_folder, model_path, csv_file_path, step=0.5):
    model = YOLO(model_path)
    video_files = [f for f in os.listdir(videos_folder) if f.endswith(('.mp4', '.MP4'))]

    for video_file in video_files:
        video_path = os.path.join(videos_folder, video_file)
        print(f"Обработка {video_path}...")
        process_video(video_path, model, csv_file_path, step=step)
        print(f"Заканчивается обработка {video_path}")
```

```python
root_videos_folder = 'ITMO-HSE-MLBD-LW-3'
model_path = 'Yolo_project_200_epoch/target_track/weights/best.pt'
csv_file_path = './result.csv'
step = 0.5

# Если CSV файл уже существует, удаляем его
if os.path.exists(csv_file_path):
    os.remove(csv_file_path)

# Получаем список всех подкаталогов
date_folders = [os.path.join(root_videos_folder, d) for d in os.listdir(root_videos_folder)
                if os.path.isdir(os.path.join(root_videos_folder, d))]

# Проходим по каждому подкаталогу с датой
for videos_folder in date_folders:
    print(f"Обработка видео в папке: {videos_folder}")
    process_all_videos(videos_folder, model_path, csv_file_path, step=step)
```

```python
def calculate_max_sequence(group):
    max_sequence = 0
    current_sequence = 0

    for hits_count, hit_in_start, hit_in_end in zip(group['hits_count'], group['hit_in_start'], group['hit_in_end']):
        if hits_count == 1:
            if hit_in_start == 1 and hit_in_end == 1:
                current_sequence += 1

            elif hit_in_start == 1 and hit_in_end == 0:
                current_sequence += 1
                max_sequence = max(max_sequence, current_sequence)
                current_sequence = 0  # обрываем после этого видео
                
            elif hit_in_start == 0 and hit_in_end == 1:
                max_sequence = max(max_sequence, current_sequence)
                current_sequence = 1  # начинаем новую последовательность
            
        elif hits_count == 2:
                current_sequence += 2

        else:
            max_sequence = max(max_sequence, current_sequence)
            current_sequence = 0

        max_sequence = max(max_sequence, current_sequence)

    return max_sequence
```
